```python
import numpy as np
from numpy.random import rand as rand
from numpy.random import seed as seed
import matplotlib.pyplot as plt
from IPython.display import clear_output
import linear_regression as LR
from misc.utils import MyUtils
import math
import sys
import pandas as pd
```

```python
# import housing data
df_X_train = pd.read_csv('houseprice/x_train.csv', header=None)
df_y_train = pd.read_csv('houseprice/y_train.csv', header=None)
df_X_test = pd.read_csv('houseprice/x_test.csv', header=None)
df_y_test = pd.read_csv('houseprice/y_test.csv', header=None)

X_train = df_X_train.to_numpy()
X_test = df_X_test.to_numpy()
n_train = X_train.shape[0]

X_all = MyUtils.normalize_0_1(np.concatenate((X_train, X_test), axis=0))
X_train = X_all[:n_train]
X_test = X_all[n_train:]

y_train = df_y_train.to_numpy()
y_test = df_y_test.to_numpy()


print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
# Closed form parameter tuning
# Will tune params in the following order
# 1) degree space = 1
    # a. lambda vals
# 2) degree space = 2
    # a. lambda vals
# 3) degree space = 3
    # a. lambda vals
# 4) degree space = 4
    # a. lambda vals

```

```python
# degree space = 1
error_arr1 = np.zeros(201)
error_arr2 = np.zeros(201)
lam_arr = np.arange(0, 2.01, 0.01)
z_r = 1
lr = LR.LinearRegression()

for i in range(201):

    # get new w vector with new lam value  
    lr.fit(X_train, y_train, CF = True, lam = i/100, eta = 0.01, epochs = 1000000, degree = z_r)
    
    error_arr1[i] = lr.error(X_train, y_train)
    error_arr2[i] = lr.error(X_test, y_test)

#print(lam_arr)
#print(error_arr)
    
plt.plot(lam_arr, error_arr1, label = "In-sample error")
plt.plot(lam_arr, error_arr2, label = "Out-of-sample error")
plt.xlabel('Lambda Values')
plt.ylabel('Error')
plt.title('Lambda Values vs Error')
plt.xlim(-0.01, 2)
#plt.ylim(1, 25)
plt.legend()
plt.show()
```

```python
# Gradient descent parameter tuning
# Will tune params in the following order
# 1) degree space = 1
    # a. lambda vals
    # b. eta 
    # c. epochs
# 2) degree space = 2
    # a. lambda vals
    # b. eta 
    # c. epochs
# 3) degree space = 3
    # a. lambda vals
    # b. eta 
    # c. epochs
# 4) degree space = 4
    # a. lambda vals
    # b. eta 
    # c. epochs
```

```python
# 1a -> lambda vals

error_arr1 = np.zeros(201)
error_arr2 = np.zeros(201)
lam_arr = np.arange(0, 2.01, 0.01)
z_r = 1
lr = LR.LinearRegression()

for i in range(201):

    # get new w vector with new lam value  
    # holding degree space, eta, and epochs constant 
    # with degree space = 1, eta = 0.01, epochs = 100000
    lr.fit(X_train, y_train, CF = False, lam = i/100, eta = 0.01, epochs = 100000, degree = z_r)
    
    error_arr1[i] = lr.error(X_train, y_train)
    error_arr2[i] = lr.error(X_test, y_test)

#print(lam_arr)
#print(error_arr)
    
plt.plot(lam_arr, error_arr1, label = "In-sample error")
plt.plot(lam_arr, error_arr2, label = "Out-of-sample error")
plt.xlabel('Lambda Values')
plt.ylabel('Error')
plt.title('Lambda Values vs Error')
#plt.xlim(0, 0.3)
#plt.ylim(0, 50)
plt.legend()
plt.show()
```
